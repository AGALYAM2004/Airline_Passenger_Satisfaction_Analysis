Satisfaction Distribution:
How many customers are satisfied vs dissatisfied

```python
airline['satisfaction'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Customer Satisfaction Distribution")
plt.show()
```

Satisfaction by Travel Class:
Which class has better customer satisfaction?

```python
airline.groupby('Class')['satisfaction'].value_counts().unstack().plot(
    kind='bar'
)

plt.title("Satisfaction by Travel Class")
plt.ylabel("Number of Passengers")
plt.show()
```

Delay Impact on Satisfaction:
Are delays causing dissatisfaction?

```python
airline.groupby('satisfaction')[['Departure Delay in Minutes',
                                'Arrival Delay in Minutes']].mean().plot(
    kind='bar'
)

plt.title("Impact of Flight Delays on Satisfaction")
plt.ylabel("Average Delay Minutes")
plt.show()
```

Service Quality Rating Comparison:
Which services need improvement?

```python
service_columns=[
'Seat comfort',
'Inflight service',
'Food and drink',
'Cleanliness',
'Online boarding',
'Baggage handling'
]

airline[service_columns].mean().plot(
    kind='bar'
)

plt.title("Average Service Ratings")
plt.ylabel("Rating Score")
plt.show()
```

Customer Type vs Satisfaction:
Are loyal customers more satisfied?

```python
airline.groupby('Customer Type')['satisfaction'].value_counts().unstack().plot(
    kind='bar',
    stacked=True
)

plt.title("Customer Type Satisfaction Analysis")
plt.ylabel("Passengers")
plt.show()
```

Age Group Satisfaction:
Does age affect satisfaction?

```python
airline.groupby('Age_Group')['satisfaction'].value_counts().unstack().plot(
    kind='bar'
)

plt.title("Age Group vs Satisfaction")
plt.show()
```

Departure Delay vs Arrival Delay
Does a higher departure delay lead to a higher arrival delay?

```python
plt.figure(figsize=(8,5))

plt.scatter(
    airline['Departure Delay in Minutes'],
    airline['Arrival Delay in Minutes']
)

plt.xlabel("Departure Delay (Minutes)")
plt.ylabel("Arrival Delay (Minutes)")
plt.title("Relationship Between Departure Delay and Arrival Delay")

plt.show()
```
